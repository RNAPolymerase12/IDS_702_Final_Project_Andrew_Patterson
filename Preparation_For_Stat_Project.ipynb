```python
import numpy as np
import pandas as pd
```

```python
#Import ILI Data
ILI = pd.read_csv('ILINET.csv')
```

```python
ILI.head()
```

```python
#Need a flu season variable
ILI['Season'] = [20000 if x > 26 else 10000 for x in ILI['WEEK']]
```

```python
ILI.sample()
```

```python
ILI['Season_Year'] = ILI['Season'] + ILI['YEAR']
```

```python
ILI.head()
```

```python
ILI = ILI.replace('X', np.NaN)
```

```python
#convert X to NA
ILI['ILITOTAL'] = ILI['ILITOTAL'].astype(float)
ILI['TOTAL PATIENTS'] = ILI['TOTAL PATIENTS'].astype(float)
```

```python
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12011, value = 22010)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12012, value = 22011)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12012, value = 22011)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12013, value = 22012)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12014, value = 22013)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12015, value = 22014)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12016, value = 22015)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12017, value = 22016)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12018, value = 22017)
ILI['Season_Year'] = ILI['Season_Year'].replace(to_replace = 12019, value = 22018)
```

```python
ILI_grouped_season = ILI.drop(columns = ['WEEK','YEAR','Season','REGION TYPE'])
ILI_grouped_season = ILI_grouped_season.groupby(['Season_Year','REGION']).sum()
```

```python
ILI_grouped = ILI_grouped_season.reset_index()
```

```python
ILI_grouped.sample(10)
```
